**Converting Python list to NumPy array**

```python
import numpy as np

lst=[1,2,3,4]
new_array=np.array(lst)

print(f"Original value is {lst} and it's type is {type(lst)}")
print(f"Original value is {new_array} and it's type is {type(new_array)}")
```

**.arange - Range of Numbers**  ---->   
*Syntax :- np.arange(start, stop, step)*

```python
first10=np.arange(10)
first10natural=np.arange(1,10)
first10even=np.arange(0,10,2)

print(f"First 10 numbers:{first10}")
print("First 10 natural numbers:",first10natural)
print(f"Even numbers upto ten:{first10even}")
```

**np.zeroes**          --->        
*To create an array with Zeroes*

```python
one_dzero=np.zeros(5)
two_dzero=np.zeros((2,5))

print(one_dzero)
print("------------")
print(two_dzero)
```

**np.ones**          --->        
*To create an array with Ones*

```python
one_dones=np.ones(5)
two_dones=np.ones((2,5))

print(one_dones)
print("------------")
print(two_dones)
```

**.linspace() - Linear Space is function used to generate evenly spaced numbers over a specified range.**   ---->       
*Synatx:- np.linspace(start, stop, num=50) --> Here stop value also exists. And num is number of values (default=50)*


```python
arr = np.linspace(0, 10)
arr2=np.linspace(1,10,3)


print(arr)  # By default it gives 50 values
print("-----------")
print(arr2)
```

**eye() is a NumPy function used to create an identity matrix.**
*Syntax:- np.eye(N, M=None, k=0) N= number of rows, M=no. of columns, k=diagonal index*

```python
arr = np.eye(3)
print(arr)

print("-----------")

arr = np.eye(4, k=1)    #upper diagonal
print(arr)

print("-----------")

arr = np.eye(4, k=-1)    #lower diagonal
print(arr)
```

**np.random.rand() is a NumPy function used to generate random numbers between 0 and 1.**   --->       
*Random, Uniformly distributed, In the range [0, 1) (0 inclusive, 1 excluded)*

```python
random_num=np.random.rand()
print(random_num)       #To print a Random Number

arr = np.random.rand(5)
print(arr)       #To print a 5 Random Numbers

mat=np.random.rand(2,3)
print(mat)
```

**.random.randn()   Generates random numbers from a standard normal distribution (mean = 0, std = 1)**  
*Random, Normal distributed (Range -inf to inf)(mean=0), In the range [0, 1) (0 inclusive, 1 excluded)*


```python
np.random.randn()        # single value
np.random.randn(5)       # 1D array
np.random.randn(3, 4)    # 2D array
```

**np.random.randint() -> Generates random integers within a specified range.**

```python
np.random.randint(10)           
np.random.randint(5, 15)        
np.random.randint(1, 10, 5)    
np.random.randint(1, 10, (3,2)) 
```

**.seed()**     ---->           
Normally, random numbers change every time you run the code.
Using .seed() helps when you want consistent results (for testing, debugging, learning).

```python
np.random.seed(5)
print(np.random.rand(1,3))
```
