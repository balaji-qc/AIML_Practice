**Series**

```python
import pandas as pd
```

```python
mydata=[1,2,3]
myindex=['a','b','c']
myser=pd.Series(mydata)
myser
```

```python
myser=pd.Series(mydata,myindex)
myser
```

```python
dictionary={'a':1, 'k':11, 'z':26}
mydict=pd.Series(dictionary)
mydict
```

```python
print(myser[2])
print(mydict['a'])
print(mydict['z'])
print(mydict.keys())
```
