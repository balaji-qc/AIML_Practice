**FILLING WITH INTERPOLATION**

```python
airline_tix = {'first':100,'business':np.nan,'economy-plus':50,'economy':30}
```

```python
ser = pd.Series(airline_tix)
```

```python
ser
```

```python
#ser.interpolate() fills missing values (NaN / NA) in a Series by estimating values between known data points
ser.interpolate()
```

```python
ser = ser.copy()
ser.index = range(len(ser))   # numeric index
```

```python
ser.interpolate(method='spline', order=2)
#Have to install 'scipy' module
#output:-
#first           100.0
#business         75.0
#economy-plus     50.0
#economy          30.0
```

```python
df = pd.DataFrame(ser,columns=['Price'])
```

```python
df
```

```python
df.interpolate()
```

```python
df=df.reset_index
```

```python
df.interpolate(method='spline',order=2)
```
